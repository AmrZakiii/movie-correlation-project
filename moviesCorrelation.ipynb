```python
#importing libraries 


import numpy as np 
import pandas as pd
import seaborn as sns 

import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize']=(12,8)
```

```python
#Read data

df=pd.read_csv("movies.csv")
df
```

```python
df.head()
```

```python
#missing data
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# zero null values
df.isna().sum()
```

```python
#data types 
df.dtypes
```

```python
#changing gross and budget to int

df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')

df.dtypes
```

```python
#Create year realsed

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
#sorting by gross
df=df.sort_values(by=["gross"],inplace=False,ascending=False)
```

```python
#Checking for dups 

df.drop_duplicates()
```

```python
# hypothesis
# Budget and gross high correlation.
# company shall have high correlation 
```

```python
#scatter plot budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel("budget for film")
plt.ylabel("gross earnings")
plt.show()
```

```python
#plot budget vs gross using seaborn
#Definetly correlated (+ve correlation)
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#Let's start looking at correlation

df.corr()   #Pearson correlation default method

#budget and gross have the highest correlation  1st hypothesis is right
```

```python
correlation = df.corr()

sns.heatmap(correlation, annot=True)
plt.title("correlation matrix for num features")
plt.show()
```
